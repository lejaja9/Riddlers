```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,99)
y = x
```

```python
fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot()
plt.title("cdf of Gym Goers", fontsize = 19)
plt.plot(x,y, lw = 3)
plt.axhline(0.5, c = 'gray', linestyle = 'dashed')
plt.axvline(0.5, c = 'black', alpha = 0.7)
plt.fill_between(x, 0, y, color = 'm', alpha = 0.2)
```
